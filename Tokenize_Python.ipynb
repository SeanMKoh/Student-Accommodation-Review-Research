```python
txtList = []
number_list=[]
for i in range (95):
    txt = df.at[i,'Review']
    txt = txt.lower()
    sentList = nltk.sent_tokenize(txt)
    for j in range(len(sentList)):
        txtList.append(sentList[j])
        number_list.append(i)

number_list
txtList
```

```python
def list_to_dataframe (List1, List2):
    df_output = pd.DataFrame(list(zip(List1, List2)))
    return df_output
```

```python
df_data = list_to_dataframe(number_list, txtList)
df_data
```
