```python
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem.wordnet import WordNetLemmatizer 
import regex as re
```

```python
df= pd.read_csv("ReviewSamples.csv")
df=df[["Review"]]
df
```

```python
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
df["Review"] = df['Review'].apply(remove_punctuations)
df
```

```python
txtList = []
reviewList= []
for index, row in df.iterrows():
    sentence= df.at[index, "Review"]
    tagged_sentence= nltk.tag.pos_tag(sentence.split())
    edited_sentence=[word for word, tag in tagged_sentence if tag=='NN' or tag=='NNS']
    txtList.append(edited_sentence)
    reviewList.append(sentence)
reviewList
txtList
```

```python
def list_to_dataframe (List1, List2):
    df_output = pd.DataFrame(list(zip(List1, List2)))
    return df_output
```

```python
df_data = list_to_dataframe(reviewList, txtList)
df_data.columns=[["Review", "Nouns"]]
df_data
```

```python
df_data.to_csv(r'WordTagging.csv')
```
